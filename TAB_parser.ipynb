```python
#logic
rows_for_json = []
for i in range(len(rows)):
    d = {}
    
    PaxName = list(rows[i].values())[0][0:60].rstrip()
    PaxBirthDate = list(rows[i].values())[0][60:72].rstrip()
    DepartDate = list(rows[i].values())[0][72:84].rstrip()
    DepartTime = list(rows[i].values())[0][84:96].rstrip()
    ArrivalDate = list(rows[i].values())[0][96:108].rstrip()
    ArrivalTime = list(rows[i].values())[0][108:120].rstrip()
    FlightCodeSh = list(rows[i].values())[0][120:132].rstrip()
    From = list(rows[i].values())[0][132:138].rstrip()
    Dest = list(rows[i].values())[0][138:144].rstrip()
    Code = list(rows[i].values())[0][144:150].rstrip()
    e_Ticket = list(rows[i].values())[0][150:168].rstrip()
    TravelDoc = list(rows[i].values())[0][168:180].rstrip()
    Seat = list(rows[i].values())[0][180:186].rstrip()
    Meal = list(rows[i].values())[0][186:192].rstrip()
    TrvCls = list(rows[i].values())[0][192:198].rstrip()
    Fare = list(rows[i].values())[0][198:204].rstrip()
    Baggage = list(rows[i].values())[0][204:216].rstrip()
    PaxAdditionalInfo = list(rows[i].values())[0][216:240].rstrip()
    FF = list(rows[i].values())[0][243:276].rstrip()
    AgentInfo = list(rows[i].values())[0][276:].rstrip()

    if len(PaxName) != 0:
        d.update({'PaxName':PaxName})
    else:
        d.update({'PaxName':'N/A'})
        
    if len(PaxBirthDate) != 0:
        d.update({'PaxBirthDate':PaxBirthDate})
    else:
        d.update({'PaxBirthDate':'N/A'})
        
    if len(DepartDate) != 0:
        d.update({'DepartDate':DepartDate})
    else:
        d.update({'DepartDate':'N/A'})
    
    if len(DepartTime) != 0:
        d.update({'DepartTime':DepartTime})
    else:
        d.update({'DepartTime':'N/A'})
        
    if len(ArrivalDate) != 0:
        d.update({'ArrivalDate':ArrivalDate})
    else:
        d.update({'ArrivalDate':'N/A'})
        
    if len(ArrivalTime) != 0:
        d.update({'ArrivalTime':ArrivalTime})
    else:
        d.update({'ArrivalTime':'N/A'})
        
    if len(FlightCodeSh) != 0:
        d.update({'FlightCodeSh':FlightCodeSh})
    else:
        d.update({'FlightCodeSh':'N/A'})
        
    if len(From) != 0:
        d.update({'From':From})
    else:
        d.update({'From':'N/A'})
        
    if len(Dest) != 0:
        d.update({'Dest':Dest})
    else:
        d.update({'Dest':'N/A'})
        
    if len(Code) != 0:
        d.update({'Code':Code})
    else:
        d.update({'Code':'N/A'})
        
    if len(e_Ticket) != 0:
        d.update({'e-Ticket':e_Ticket})
    else:
        d.update({'e-Ticket':'N/A'})
        
    if len(TravelDoc) != 0:
        d.update({'TravelDoc':TravelDoc})
    else:
        d.update({'TravelDoc':'N/A'})
        
    if len(Seat) != 0:
        d.update({'Seat':Seat})
    else:
        d.update({'Seat':'N/A'})
        
    if len(Meal) != 0:
        d.update({'Meal':Meal})
    else:
        d.update({'Meal':'N/A'})
        
    if len(TrvCls) != 0:
        d.update({'TrvCls':TrvCls})
    else:
        d.update({'TrvCls':'N/A'})
        
    if len(Fare) != 0:
        d.update({'Fare':Fare})
    else:
        d.update({'Fare':'N/A'})
        
    if len(Baggage) != 0:
        d.update({'Baggage':Baggage})
    else:
        d.update({'Baggage':'N/A'})
        
    if len(PaxAdditionalInfo) != 0:
        d.update({'PaxAdditionalInfo':PaxAdditionalInfo})
    else:
        d.update({'PaxAdditionalInfo':'N/A'})
        
    if len(FF) != 0:
        d.update({'FF':FF})
    else:
        d.update({'FF':'N/A'})
        
    if len(AgentInfo) != 0:
        d.update({'AgentInfo':AgentInfo})
    else:
        d.update({'AgentInfo':'N/A'})

    rows_for_json.append(d)
```
